```python
import os
import pandas as pd
import numpy as np

frames = 100
# Default value for empty cells
default = .5

modified = "./data/absolute/modified"
os.mkdir(modified)

dirname = "./data/absolute/2D/" 

#Preparation Stage - Load data and normalize
listfile = os.listdir(dirname)
listfile= sorted(listfile, key=str.casefold) 
data = []
data_new = []

for wordname in listfile:
        if wordname == ".DS_Store":
            continue
        for csv in os.listdir(dirname + wordname):
            filepath = os.path.join(dirname, wordname, csv)
            print(filepath)
            content = pd.read_csv(filepath, sep=';')
            content = content + 0.1
            new_filepath = os.path.join(modified, wordname, "modified_"+csv)
            new_filedir = os.path.join(modified, wordname )
            if not os.path.exists(new_filedir):
                os.mkdir(new_filedir)
            content.to_csv(new_filepath, sep=';', index=False)
            
            
```
